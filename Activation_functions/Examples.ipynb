# Sigmoid

$\text{Formula: }$<p>
$$f(x)=\frac{1}{1 + e^{-x}}$$
$\text{Derivative: }$<p>
$$\frac{\partial f(x)}{\partial x}=f(x)\times(1-f(x))$$


```python
sigmoid_function = Sigmoid()
y = sigmoid_function(x)
fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', line=dict(color="blue"), showlegend=False))
fig.update_layout(template="simple_white", width=800, height=800, title="<b>Sigmoid function<b>", title_x=0.5, xaxis_title="x", yaxis_title="y", font=dict(family="Times New Roman",size=16,color="Black"))
fig.show("png")
```

```python
derivative = sigmoid_function.derivative(x)
fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=derivative, mode='lines', line=dict(color="blue"), showlegend=False))
fig.update_layout(template="simple_white", width=800, height=800, title="<b>Derivative of Sigmoid function<b>", title_x=0.5, xaxis_title="x", yaxis_title="y", font=dict(family="Times New Roman",size=16,color="Black"))
fig.show("png")
```

```python
weights = torch.tensor([0.5, 1, 5, 10])
colors = ["blue", "red", "green", "purple"]
fig = go.Figure()
for idx, weight in enumerate(weights):
    y = sigmoid_function(x*weight)
    fig.add_trace(go.Scatter(x=x, y=y, mode='lines', line=dict(color=colors[idx], width=2), name=f"Weight: {weight}"))
fig.update_layout(template="simple_white", width=800, height=800, title="<b>Sigmoid function for different weights<b>", title_x=0.5, xaxis_title="x", yaxis_title="y", font=dict(family="Times New Roman",size=16,color="Black"))
fig.show("png")
```

```python
bias = torch.tensor([0.5, 1, 5, 10])
colors = ["blue", "red", "green", "purple"]
fig = go.Figure()
for idx, b in enumerate(bias):
    y = sigmoid_function(x + b)
    fig.add_trace(go.Scatter(x=x, y=y, mode='lines', line=dict(color=colors[idx], width=2), name=f"Bias: {b}"))
fig.update_layout(template="simple_white", width=800, height=800, title="<b>Sigmoid function for different bias values<b>", title_x=0.5, xaxis_title="x", yaxis_title="y", font=dict(family="Times New Roman",size=16,color="Black"))
fig.show("png")
```

# Softmax

$\text{Formula: }$<p>
$$f(x_i)=\frac{e^{x_i}}{\sum_{j=1}^{n}e^{x_j}}$$
$\text{Derivative: }$<p>
$$\frac{\partial f(x_i)}{\partial x_j}=f(x_i)\times(1-f(x_j))$$


```python
softmax_function = Softmax()
y = softmax_function(x)
fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', line=dict(color="blue"), showlegend=False))
fig.update_layout(template="simple_white", width=800, height=800, title="<b>Softmax function<b>", title_x=0.5, xaxis_title="x", yaxis_title="y", font=dict(family="Times New Roman",size=16,color="Black"))
fig.show("png")
```

```python
derivative = softmax_function.derivative(x)
fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=derivative, mode='lines', line=dict(color="blue"), showlegend=False))
fig.update_layout(template="simple_white", width=800, height=800, title="<b>Derivative of Softmax function<b>", title_x=0.5, xaxis_title="x", yaxis_title="y", font=dict(family="Times New Roman",size=16,color="Black"))
fig.show("png")
```

```python
weights = torch.tensor([0.5, 1, 5, 10])
colors = ["blue", "red", "green", "purple"]
fig = go.Figure()
for idx, weight in enumerate(weights):
    y = softmax_function(x*weight)
    fig.add_trace(go.Scatter(x=x, y=y, mode='lines', line=dict(color=colors[idx], width=2), name=f"Weight: {weight}"))
fig.update_layout(template="simple_white", width=800, height=800, title="<b>Softmax function for different weights<b>", title_x=0.5, xaxis_title="x", yaxis_title="y", font=dict(family="Times New Roman",size=16,color="Black"))
fig.show("png")
```

# ReLU

$$f(x)=\begin{cases} x & \text{if } x > 0 \\ 0 & \text{otherwise} \end{cases}$$
$\text{Derivative: }$<p>
$$\frac{\partial f(x)}{\partial x}=\begin{cases} 1 & \text{if } x > 0 \\ 0 & \text{otherwise} \end{cases}$$

# Tanh

$\text{Formula: }$<p>
$$f(x)=\frac{e^x - e^{-x}}{e^x + e^{-x}}$$
$\text{Derivative: }$<p>
$$\frac{\partial f(x)}{\partial x}=1-f(x)^2$$

```python
tanh_function = Tanh()
y = tanh_function(x)
fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', line=dict(color="blue"), showlegend=False))
fig.update_layout(template="simple_white", width=800, height=800, title="<b>Tanh function<b>", title_x=0.5, xaxis_title="x", yaxis_title="y", font=dict(family="Times New Roman",size=16,color="Black"))
fig.show("png")
```

```python
derivative = tanh_function.derivative(x)
fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=derivative, mode='lines', line=dict(color="blue"), showlegend=False))
fig.update_layout(template="simple_white", width=800, height=800, title="<b>Derivative of Tanh function<b>", title_x=0.5, xaxis_title="x", yaxis_title="y", font=dict(family="Times New Roman",size=16,color="Black"))
fig.show("png")
```

```python
weights = torch.tensor([0.5, 1, 5, 10])
colors = ["blue", "red", "green", "purple"]
fig = go.Figure()
for idx, weight in enumerate(weights):
    y = tanh_function(x*weight)
    fig.add_trace(go.Scatter(x=x, y=y, mode='lines', line=dict(color=colors[idx], width=2), name=f"Weight: {weight}"))
fig.update_layout(template="simple_white", width=800, height=800, title="<b>Tanh function for different weights<b>", title_x=0.5, xaxis_title="x", yaxis_title="y", font=dict(family="Times New Roman",size=16,color="Black"))
fig.show("png")
```

# Leaky ReLU

$\text{Formula: }$<p>
$$f(x)=\begin{cases} x & \text{if } x > 0 \\ \alpha\times x & \text{otherwise} \end{cases}$$
$\text{Derivative: }$<p>
$$\frac{\partial f(x)}{\partial x}=\begin{cases} 1 & \text{if } x > 0 \\ \alpha & \text{otherwise} \end{cases}$$


```python
leaky_relu_function = LeakyReLU(alpha=0.1)
y = leaky_relu_function(x)
fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', line=dict(color="blue"), showlegend=False))
fig.update_layout(template="simple_white", width=800, height=800, title="<b>Leaky ReLU function<b>", title_x=0.5, xaxis_title="x", yaxis_title="y", font=dict(family="Times New Roman",size=16,color="Black"))
fig.show("png")
```

```python
weights = torch.tensor([0.5, 1, 5, 10])
colors = ["blue", "red", "green", "purple"]
fig = go.Figure()
for idx, weight in enumerate(weights):
    y = leaky_relu_function(x*weight)
    fig.add_trace(go.Scatter(x=x, y=y, mode='lines', line=dict(color=colors[idx], width=2), name=f"Weight: {weight}"))
fig.update_layout(template="simple_white", width=800, height=800, title="<b>Leaky ReLU function for different weights<b>", title_x=0.5, xaxis_title="x", yaxis_title="y", font=dict(family="Times New Roman",size=16,color="Black"))
fig.show("png")
```

# ELU

$\text{Formula: }$<p>
$$f(x)=\begin{cases} x & \text{if } x > 0 \\ \alpha\times(e^x-1) & \text{otherwise} \end{cases}$$
$\text{Derivative: }$<p>
$$\frac{\partial f(x)}{\partial x}=\begin{cases} 1 & \text{if } x > 0 \\ f(x)+\alpha & \text{otherwise} \end{cases}$$


```python
elu_function = ELU(alpha=0.1)
y = elu_function(x)
fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', line=dict(color="blue"), showlegend=False))
fig.update_layout(template="simple_white", width=800, height=800, title="<b>ELU function<b>", title_x=0.5, xaxis_title="x", yaxis_title="y", font=dict(family="Times New Roman",size=16,color="Black"))
fig.show("png")
```

```python
derivative = elu_function.derivative(x)
fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=derivative, mode='lines', line=dict(color="blue"), showlegend=False))
fig.update_layout(template="simple_white", width=800, height=800, title="<b>Derivative of ELU function<b>", title_x=0.5, xaxis_title="x", yaxis_title="y", font=dict(family="Times New Roman",size=16,color="Black"))
fig.show("png")
```

```python
weights = torch.tensor([0.5, 1, 5, 10])
colors = ["blue", "red", "green", "purple"]
fig = go.Figure()
for idx, weight in enumerate(weights):
    y = elu_function(x*weight)
    fig.add_trace(go.Scatter(x=x, y=y, mode='lines', line=dict(color=colors[idx], width=2), name=f"Weight: {weight}"))
fig.update_layout(template="simple_white", width=800, height=800, title="<b>ELU function for different weights<b>", title_x=0.5, xaxis_title="x", yaxis_title="y", font=dict(family="Times New Roman",size=16,color="Black"))
fig.show("png")
```
